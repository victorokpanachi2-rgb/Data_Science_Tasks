##**TASK 1 - SPORTS PLAYER PERFORMANCE ANALYSIS**


##Question 1: Store the values in a Python list.

##Solution:


```python
goals = [12, 7, 9, 15, 6, 10]  # 1. Player's goals scored in different matches
```

##Question 2: Calculate and print the total number of scored.

##Solution:


```python
sum_of_goals = sum(goals)  # 2. Calculate the total number of goals scored
print("Total number of goals scored:", sum_of_goals)
```

##Question 3: Find and print the difference between the highest and lowest goals scored in a match.

##Solution:

```python
highest_goals = max(goals)
lowest_goals = min(goals)

difference = highest_goals - lowest_goals  # 3. Find the difference between the highest and lowest goals scored
print("Difference between highest and lowest goals:", difference)
```

##Question 4: Calculate and print the average goals per match.

##Solution:

```python
average_goals =  sum_of_goals/ len(goals)  # 4. Calculate the average goals per match
print("Average goals per match:", average_goals)
```

##Question 5: Add a new match where the player scored 11 goals.

##Solution:

```python
goals.append(11)  # 5. Add a new match where the player scored 11 goals
print("After adding new match:", goals)
```

##Question 6: Remove the lowest-scoring match from the list.

##Solution:

```python
goals.remove(min(goals))  # 6. Remove the lowest-scoring match
print("After removing the lowest score:", goals)
```

##Question 7: Print the final updated list of performance data.

##Solution:

```python
print("Final updated performance data:", goals)  # 7. Print the final updated list
```

##**TASK 3 - CPU TEMPERATURE MONITORING SYSTEM**

##Question 1: Store the values in a Python tuple.
##Solution:

```python
temperatures = (45, 50, 55, 48, 52)  #1 CPU temperature readings stored in a tuple
```

##Question 2: Print the first temperature reading.

##Solution:

```python
print("First temperature reading:", temperatures[0])  # 2. Print the first temperature reading
```

##Question 3: Print the last temperature reading.

##Solution:

```python
print("Last temperature reading:", temperatures[-1]) # 3. Print the last temperature reading
```

##Question 4: Find and print the difference between the highest and lowest temperatures.

##Solution:

```python
highest_temp = max(temperatures)
lowest_temp = min(temperatures)

difference = highest_temp - lowest_temp  # 4. Find the difference between the highest and lowest temperatures
print("Difference between highest and lowest temperatures:", difference)
```

##Question 5: Calculate and print the average CPU temperature.

##Solution:

```python
average_temp = sum(temperatures) / len(temperatures) # 5. Calculate the average CPU temperature
print("Average CPU temperature:", average_temp)
```

##Question 6: In 1-2 sentences, explain why a tuple is more suitable than a list for storing CPU temperature data.

##Solution:

A tuple is more suitable because CPU temperature readings are not meant to be changed accidentally during program execution. Tuples are immutable, which makes the data more secure and reliable compared to lists.
